```python
#most influential data is Absolute magnitude and Radius
df = pd.read_csv('/kaggle/input/star-dataset/6 class csv.csv')
#df.query("Star type")
a=0
Type1R =[]
Type1T=[]
Type1A=[]
b=0
Type2R =[]
Type2T=[]
Type2A=[]
c=0
Type3R =[]
Type3T=[]
Type3A=[]
d=0
Type4R =[]
Type4T=[]
Type4A=[]
e=0
Type5R =[]
Type5T=[]
Type5A=[]
f=0
Type6R =[]
Type6T=[]
Type6A=[]
count=0
for r in df['Star type']:
    data=df['Absolute magnitude(Mv)'][count]
    data2=df['Temperature (K)'][count]
    
    

    if r==0:
        Type1R.append(df['Radius(R/Ro)'][count])
        Type1A.append(data)
        Type1T.append(data2)
        a=a+1
    if r == 1:
        Type2R.append(df['Radius(R/Ro)'][count])
        Type2A.append(data)
        Type2T.append(data2)
        b=b+1
    if r==2:
        Type3R.append(df['Radius(R/Ro)'][count])
        Type3A.append(data)
        Type3T.append(data2)
        c=c+1
    if r==3:
        Type4R.append(df['Radius(R/Ro)'][count])
        Type4A.append(data)
        Type4T.append(data2)
        d=d+1
    if r==4:
        Type5R.append(df['Radius(R/Ro)'][count])
        Type5A.append(data)
        Type5T.append(data2)
        e=e+1
    if r==5:
        Type6R.append(df['Radius(R/Ro)'][count])
        Type6A.append(data)
        Type6T.append(data2)
        f=f+1
    count=count+1

datR=df['Radius(R/Ro)'] 
datL=df['Luminosity(L/Lo)']
datT=df['Temperature (K)']
datAM=df['Absolute magnitude(Mv)']
datSC=df['Spectral Class']
datC=df['Star color']


    
rAVG=np.mean(Type1R)
rSTD1=np.std(Type1R)
rSTD2=np.std(Type2R)
rSTD3=np.std(Type3R)
rSTD4=np.std(Type4R)
rSTD5=np.std(Type5R)
rSTD6=np.std(Type6R)

print(rSTD1)
print(rSTD2)
print(rSTD3)
print(rSTD4)
print(rSTD5)
print(rSTD6)
#standard deviation of radius correlates to Star Type


data = [Type1R,Type2R,Type3R]


fig = plt.figure(figsize =(10, 7))
ax = fig.add_axes([0, 0,1,1])
bp = ax.boxplot(data)

plt.plot()
plt.xlabel("Star Type(1,2,3)")
plt.ylabel("radius")


#box and whisker plot displaying how the radius differs for different Star types
#our sun data: 5778K 4.83(Mv ) Yellow class:G 
```
